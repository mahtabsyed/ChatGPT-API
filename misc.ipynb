```python
import os
from PyPDF2 import PdfReader
from docx import Document

# Set the folder path where the files are located
folder_path = "/Users/mahtabsyed/Library/Mobile Documents/com~apple~CloudDocs/All Family Storage/1 - Career - Mahtab/UST/Hiring/Data Engineer (Del Davies, Macquarie)"

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    # Check if the file is a pdf document
    if filename.endswith(".pdf"):
        with open(file_path, "rb") as pdf_file:
            pdf_reader = PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)

            # Extract text from each page
            for page_num in range(num_pages):
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                print(text)

    # Check if the file is a Word document
    elif filename.endswith(".docx"):
        # Create a Word document object
        doc = Document(file_path)

        # Get the first paragraph
        paragraph = doc.paragraphs[0]

        # Extract the text from the first paragraph
        text = paragraph.text

        # Split the text into sentences
        sentences = text.split(".")

        # Print the file name and the first sentence
        print(f"File: {filename}")
        print(f"First sentence: {sentences[0] + '.'}")
```
